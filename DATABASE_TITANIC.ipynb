```python
import pandas as pd
from seaborn import load_dataset

df = load_dataset('titanic')

df.shape
```

```python
df = df.dropna()
df.shape
df
```

```python
df = df.drop(columns=['class'])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer


column_transformer = make_column_transformer ((OneHotEncoder(), ['sex','who', 'adult_male', 'embark_town','alone', 'alive']), remainder = 'passthrough')
df = column_transformer.fit_transform(df)


```

```python
df = pd.DataFrame(data=df, columns=column_transformer.get_feature_names_out())
```

```python
df
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
df['remainder__embarked'] = label_encoder.fit_transform(df['remainder__embarked'])
df['remainder__deck'] = label_encoder.fit_transform(df['remainder__deck'])
df['remainder__pclass'] = label_encoder.fit_transform(df['remainder__pclass'])
```

```python
df
```

```python
y = df['remainder__pclass']

x = df.drop('remainder__pclass', axis = 1)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,  test_size=0.2, random_state=None, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

Arvore1 = DecisionTreeClassifier(criterion = "entropy")
Arvore2 = DecisionTreeClassifier(criterion = "entropy" ,max_depth= 2)
Knn1 = KNeighborsClassifier(n_neighbors=7, metric = 'euclidean', algorithm = 'brute')
Knn2 = KNeighborsClassifier(n_neighbors= 3, metric = 'euclidean', algorithm = 'brute')

Arvore1.fit(x_train, y_train)
Arvore2.fit(x_train, y_train)
Knn1.fit(x_train, y_train)
Knn2.fit(x_train, y_train)

```

```python
from sklearn import metrics

result1 = Arvore1.predict(x_test)
result2 = Arvore2.predict(x_test)
result3 = Knn1.predict(x_test)
result4 = Knn1.predict(x_test)



acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1*100)
print("{}%".format(show1))

acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2*100)
print("{}%".format(show2))

acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3*100)
print("{}%".format(show3))

acc4 = metrics.accuracy_score(result4, y_test)
show4 = round(acc4*100)
print("{}%".format(show4))
```
