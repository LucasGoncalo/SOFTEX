```python
import pandas as pd # Pandas é utilizado para o tratamento de dados
from seaborn import load_dataset

df = load_dataset('penguins') # Data base que identifica o tipo de pinguins

df
```

```python
df.info() # Mostra informações do data base
df.shape # Mostra a quantidade de linhas e colunas
df.describe() # Faças uma média dos valores inteiros no código
```

```python
df = df.dropna() #Dropa o número de linha com valores vázios
df.shape
```

One-Hot Encoder

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

column_transformer = make_column_transformer((OneHotEncoder(), ['island', 'sex']), remainder = 'passthrough')

df = column_transformer.fit_transform(df) # fit_transform(df) Transforma as colunas de classificação em números, aplicando VDM
# OneHotEncoderé utilizados para características, tipo sexo ou país
df = pd.DataFrame(data=df)
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder() # Já o labelencoder é utilizado para classe, ou seja, se aquele pinguim é da mesma "classe". cada classe recebe apenas um número

df[5] = label_encoder.fit_transform(df[5])

```

```python
y = df[5]

x = df.drop(5, axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,  test_size=0.2, random_state=3, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

model1 = DecisionTreeClassifier(criterion = "entropy")
model2 = KNeighborsClassifier(n_neighbors=7, metric = 'euclidean', algorithm = 'brute')
model3 = DecisionTreeClassifier(criterion = "entropy", max_depth=1)

model1.fit(x_train, y_train)
model2.fit(x_train, y_train)
model3.fit(x_train, y_train)
```

```python
from sklearn import metrics

result1 = model1.predict(x_test)
result2 = model2.predict(x_test)
result3 = model3.predict(x_test)

acc1 = metrics.accuracy_score(result1, y_test)
show1 = round(acc1*100)
print("{}%".format(show1))

acc2 = metrics.accuracy_score(result2, y_test)
show2 = round(acc2*100)
print("{}%".format(show2))

acc3 = metrics.accuracy_score(result3, y_test)
show3 = round(acc3*100)
print("{}%".format(show3))
```
